# 1. Categorical Data

a. Countplot

```python
import seaborn as sns
```

```python
# sns.countplot(df['Survived'])
sns.countplot(x='Survived', data=df)
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

Here the data is catagorical because it has two catagories survived or not.

```python
sns.countplot(x='Pclass', data=df)  #this shows in which class did the people travelled
```

```python
sns.countplot(x='Sex', data=df)  #sex of the passenger travelling in titanic
```

```python
sns.countplot(x='Embarked', data=df)  #From which port did the passengers mos tof them onboarded in the ship
```

# b. To get in percentages

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
```
